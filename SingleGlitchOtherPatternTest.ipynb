# Define Noise

```python
#Hopefully this will be more advanced in future. Allowing you to specify the noise level in snr.
def getNoise(signal,stdDev=0):
    if stdDev == 0:
        noise =  np.zeros(len(signal))
    else:
        noise =  np.random.normal(0,stdDev,len(signal))
    return noise
```

```python
def proximalSynapseMagnitude(model,result):
    p=r.fastProximalPermanences(model,None)
    return np.sqrt(p.dot(p))
```

```python
def trainModelOnGlitch1WithNoise(trainingNoiseLevel,trainData,preexistingModel=None,repititions=200,fieldFunctions=[proximalSynapseMagnitude]):
    #Initialize the model
    if preexistingModel==None:
        model = ModelFactory.create(MODEL_PARAMS)
        model.enableInference({"predictedField": "sine"})
    else:
        model = preexistingModel

    #Initialise data structure
    try:
        trainData[trainingNoiseLevel]
    except:
        trainData[trainingNoiseLevel] = r.Store()
    
    #Train the model
    for i in xrange(repititions):
 
        for sample in trimmedGlitch1+getNoise(trimmedGlitch1,trainingNoiseLevel):
            result = model.run({"sine": float(sample)})
            trainData[trainingNoiseLevel].record(model,result,fieldFunctions)
        
        #reset inbetween each signal
        model.resetSequenceStates()
    return model
#r.modelInput,r.bestPrediction,r.anomalyScore,
```
