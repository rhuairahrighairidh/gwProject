# Define Noise

```python
#Hopefully this will be more advanced in future. Allowing you to specify the noise level in snr.
def getNoise(signal,stdDev=0):
    if stdDev == 0:
        noise =  np.zeros(len(signal))
    else:
        noise =  np.random.normal(0,stdDev,len(signal))
    return noise
```

# Train Model On Glitch1 + Noise

```python
#Initialize the model and data structure.
model = ModelFactory.create(MODEL_PARAMS)
model.enableInference({"predictedField": "sine"})
trainData = {}
```

```python
trainingNoiseLevel = 0.1 #set the stdDev of noise to add to glitch1 when training

trainData[trainingNoiseLevel] = {}
trainData[trainingNoiseLevel]['actual'] = []
trainData[trainingNoiseLevel]['anomalys'] = []
trainData[trainingNoiseLevel]['predicted'] = []
trainData[trainingNoiseLevel]['predictedColumns'] = []
```

Train on glitch1 repeatedly.

```python
for i in xrange(600):
 
    for sample in glitch1+getNoise(glitch1,trainingNoiseLevel):
        result = model.run({"sine": float(sample)})
        trainData[trainingNoiseLevel]['actual'].append(sample)
        trainData[trainingNoiseLevel]['anomalys'].append(result.inferences['anomalyScore'])
        trainData[trainingNoiseLevel]['predicted'].append(result.inferences['multiStepBestPredictions'][1])
        trainData[trainingNoiseLevel]['predictedColumns'].append(model._prevPredictedColumns.shape[0])
        
    #reset inbetween each signal
    model.resetSequenceStates()
```
