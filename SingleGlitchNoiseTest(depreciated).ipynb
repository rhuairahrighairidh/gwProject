# Define Noise

````python
#Hopefully this will be more advanced in future. Allowing you to specify the noise level in snr.
def getNoise(signal,stdDev=0):
    if stdDev == 0:
        noise =  np.zeros(len(signal))
    else:
        noise =  np.random.normal(0,stdDev,len(signal))
    return noise
````

The probability distribution of a predicted value can be multi modal.

````python
def runTest(model,testSignal):
    store = {}
    store['actual']=[]
    store['predicted']=[]
    store['anomalys']=[]
    
    for sample in testSignal:
        result = model.run({"sine": float(sample)})
        store['actual'].append(sample)
        store['anomalys'].append(result.inferences['anomalyScore'])
        store['predicted'].append(result.inferences['multiStepBestPredictions'][1])
        
    return store
````

````python
testData={}
````

````python
##test model on glitch with various levels of noise, store all data generated in a big dictionary
numberOfRepetitions = 100 #The higher the better
model.disableLearning()

for testingNoiseLevel in [float(i)/20 for i in range(0,11)]:
    testData[testingNoiseLevel]={}
    for i in xrange(numberOfRepetitions):
        model.resetSequenceStates()
        testData[testingNoiseLevel][i] = runTest(model,glitch1+getNoise(glitch1,testingNoiseLevel))
````

````python
##calculate average error for each noise level, using the big dictionary from above
def computeError(dictionaryOfData): #Error bars would be nice as well
    actual = np.array(dictionaryOfData['actual'][1:])
    predicted = np.array(dictionaryOfData['predicted'][:-1])
    return np.sum(np.abs(actual-predicted)) / len(predicted)
````

````python
#Calculate Errors
errors[trainingNoiseLevel] = {}
errorBars[trainingNoiseLevel] = {}
for testNoiseLevel in testData.keys():
    errors[trainingNoiseLevel][testNoiseLevel] = np.mean(map(computeError,testData[testNoiseLevel].values()))
    errorBars[trainingNoiseLevel][testNoiseLevel] = np.std(map(computeError,testData[testNoiseLevel].values()))
````
