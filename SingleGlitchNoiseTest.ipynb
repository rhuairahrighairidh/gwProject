# Define Noise

```python
#Hopefully this will be more advanced in future. Allowing you to specify the noise level in snr.
def getNoise(signal,stdDev=0):
    if stdDev == 0:
        noise =  np.zeros(len(signal))
    else:
        noise =  np.random.normal(0,stdDev,len(signal))
    return noise
```

```python
def trainModelOnGlitch1WithNoise(trainingNoiseLevel,trainData):
    #Initialize the model
    model = ModelFactory.create(MODEL_PARAMS)
    model.enableInference({"predictedField": "sine"})

    #Initialise data structure
    trainData[trainingNoiseLevel] = r.Store()
    
    #Train the model
    for i in xrange(200):
 
        for sample in glitch1+getNoise(glitch1,trainingNoiseLevel):
            result = model.run({"sine": float(sample)})
            trainData[trainingNoiseLevel].record(model,result,[r.modelInput,r.bestPrediction,r.anomalyScore])
        #reset inbetween each signal
        model.resetSequenceStates()
        
    return model
```

# Initialize Functions Used For Testing

```python
from NoiseTestFunctions import *
```

These functions below now reside in `NoiseTestFunctions.py`

```python
def runModel(model,testSignal,fieldFunctions = [r.modelInput,r.bestPrediction,r.anomalyScore]):
    """
    This runs the model through the supplied signal. At each point it saves out data specied in fieldFunctions.
    This takes the model input to be the first field in model.getFieldInfo().
    """
    store = r.Store()
    
    for sample in testSignal:
        result = model.run({model.getFieldInfo()[0].name: float(sample)})
        store.record(model,result,fieldFunctions)
  
    return store
```

```python
##calculate average error for each noise level, using the big dictionary from above
def computeError(store): #Error bars would be nice as well
    actual = np.array(store.modelInput[1:])
    predicted = np.array(store.bestPrediction[:-1])
    return np.sum(np.abs(actual-predicted)) / len(predicted)
```

```python
def computeAverageAnomaly(store):
    return np.array(store.anomalyScore).mean()
```

```python
def runAverageingTest(model,testSignals,errorFunction=computeError):
    """
    This function runs the model on each signal in the list testSignals, reseting the sequence inbetween each.
    For each sigal an error is computed (using the suplied errorFunction).
    The average and standard deviation of these errors will be returned.
    """
    errors = []
    for signal in testSignals:
        model.resetSequenceStates()
        errors.append(errorFunction(runModel(model,signal)))#The data generated when running is lost here. You could modify this to save it into a store.
    
    errors = np.array(errors)
    return {'average': errors.mean(), 'stdDev':errors.std()}
```

# Attempting To Reduce Prediction Error By Training With Noise

Rerun the training and testing code above but with several different values for noise stdDev.

This will result in an errors dictionary with all the right information.

```python
#These values were used for training
trainingNoiseLevels=[float(i)/20 for i in range(1,11,2)]
trainingNoiseLevels
```

```python
allErrors={}
for noiseLevel in trainingNoiseLevels:
    model=trainModelOnGlitch1WithNoise(noiseLevel,trainData) #Store the data in the dictionary used at the start

    ##test model on glitch with various levels of noise
    allErrors[noiseLevel]={}
    glitch1Errors = allErrors[noiseLevel] #This doesn't copy the dictionary
    model.disableLearning()
    for testingNoiseLevel in [float(i)/20 for i in range(0,11)]:
        testSignals = [glitch1+getNoise(glitch1,testingNoiseLevel) for i in range(100)] #This defines the number of repitions to average over.
        glitch1Errors[testingNoiseLevel] = runAverageingTest(model,testSignals, errorFunction=computeError)
```

```python
(fig,ax)=plt.subplots(1,1)
for noiseLevel in trainingNoiseLevels:
    x=sorted(allErrors[noiseLevel].keys())
    y=[allErrors[noiseLevel][n]['average'] for n in x]
    er=[allErrors[noiseLevel][n]['stdDev'] for n in x]
    ax.errorbar(x,y,marker='x',linestyle='-',yerr=er,label=noiseLevel)
ax.set_xlim(left=-0.01,right=0.51)
ax.set_ylim(bottom=-0.01,top = 0.65)
ax.legend(loc=0)
ax.set_ylabel('Average Error')
ax.set_xlabel('Standadrd Deviation of Noise Added In Testing')
fig.set_size_inches([10,7])
```
