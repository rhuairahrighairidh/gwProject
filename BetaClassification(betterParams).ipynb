# Encode All The Beta Values

```python
def encode(inputScalar, ScalarEncoderInstance=se):
    array = np.zeros(ScalarEncoderInstance.getWidth())
    ScalarEncoderInstance.encodeIntoArray(inputScalar,array)
    return array
```

```python
print encode(0.1)
```

Encode each beta value into an array, then join all the arrays (corresponding to one vector) together into a 2D array. This will be the inputs to be fed to the spatial pooler.
In this case the input array doesn't need to be 2D. The individual arrays could have been joined end to end.

```python
def convertBetas(Betas):
    encodedBetas = []
    for betaVector in Betas:
        encodedBetas.append(map(encode,betaVector))
    encodedBetas = map(np.vstack,encodedBetas)
    return encodedBetas
```

```python
encodedG = convertBetas(G)
encodedSG = convertBetas(SG)
encodedRD = convertBetas(RD)
```

# Set Up Spatial Pooler

Initialize a spatial pooler with some parameters. Again these are optimum settings.

```python
sp = SpatialPooler(inputDimensions=(10, 223),
                   columnDimensions=(32, 64),
                   potentialRadius=10,
                   potentialPct=0.7,
                   globalInhibition=True,
                   localAreaDensity=-1.0,
                   numActiveColumnsPerInhArea=40,
                   stimulusThreshold=0,
                   synPermInactiveDec=0.01,
                   synPermActiveInc=0.1,
                   synPermConnected=0.1,
                   minPctOverlapDutyCycle=0.001,
                   minPctActiveDutyCycle=0.001,
                   dutyCyclePeriod=1000,
                   maxBoost=10.0,
                   seed=-1,
                   spVerbosity=0)
```

Initialise data recording. (or don't, this slows everything down 10 times or more)

```python
Patcher('run1').patchSP(sp)
```

### Training (feed in the first 80% of the data)

Here the encoded version of each beta vector is fed into the spatial pooler one after another. Ideally the spatial pooler gradually learns how to best represent the input.

```python
outputArray = np.zeros(32*64)
for i in xrange(100,600):
    for inputArray in (encodedG[i],encodedSG[i],encodedRD[i]):
        sp.compute(inputArray.flatten(), True, outputArray)
```

### Testing (see what SDRs you get when you feed in the rest of the data)

Feed in the rest of the data and store the resulting SDRs, seperating them into lists according to the original class of beta.

Here the learning has been turned off.

```python
def getIndexes(sdr):
    return np.where(sdr)[0]
```

```python
allOutput=[]
for i in xrange(600,750):
    for inputArray in (encodedG[i],encodedSG[i],encodedRD[i]):
        outputArray = np.zeros(32*64)
        sp.compute(inputArray.flatten(), False, outputArray) #learning turned off
        allOutput.append(getIndexes(outputArray))
#extract the SDRs corresponding to different inputs into different lists
GOutput = allOutput[0::3]
SGOutput = allOutput[1::3]
RDOutput = allOutput[2::3]
```
