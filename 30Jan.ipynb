```python
from IPython.display import clear_output
```

```python
anomalys = []
predicted = []
counter = 0
for reading in data:
    result = model.run({"volts": float(reading)})
    anomalys.append(result.inferences['anomalyScore'])
    predicted.append(result.inferences['multiStepBestPredictions'][1])
    counter +=1
    clear_output(wait=True)
    print counter
```

```python
lines=np.concatenate(([data],10*np.array([anomalys]),np.array([predicted])),axis=0).T
```
