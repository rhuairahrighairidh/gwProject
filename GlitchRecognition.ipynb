`ModelFactory.create(MODEL_PARAMS)`
 - [in file `frameworks/opf/modelfactory.py`]
 - `MODEL_PARAMS['model']` specifies model type to be created. It can be a "CLA", "TwoGram" or "PreviousValue" model.
 - Creates a CLA model
 - returns `CLAModel(**MODEL_PARAMS['modelParams'])` passes value of the key 'modelParams' to `__init__` in `CLAModel`

CLAModel class in file `frameworks/opf/clamodel.py`
 - InferenceType.TemporalNextStep
 - InferenceType.TemporalClassification
 - InferenceType.NontemporalClassification
 - InferenceType.NontemporalAnomaly
 - InferenceType.TemporalAnomaly
 - InferenceType.TemporalMultiStep
 - InferenceType.NontemporalMultiStep
 - imports `Anomaly` from `nupic.algorithms.anomaly`
 - The init has several arguments. One is `anomalyParams`
 - if inference type is non temporal then TP isn't created
 - `self._anomalyInst = Anomaly(slidingWindowSize=windowSize, mode=mode)` both args come straight from `anomalyParams` (if their not there then args default to `None` and `"pure"` -> this used in method `self._anomalyCompute` -> which is used once in `self.run()`
 - `__createCLANetwork` also gets passed `anomalyParams` -> This does `addAnomalyClassifierRegion()` only if it's `TemporalAnomaly`
 - NontemporalAnomaly => set anomalyMode of SP True
 - TemporalAnomaly => set anomalyMode of TP True
